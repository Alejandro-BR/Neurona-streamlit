# Ejercicio 3 Una neurona con tres entradas y bias (sesgo)

```python
text = "Introduzca el valor de las tres entradas separadas por un espacio: "
x1, x2, x3 = map(float ,input(text).split())

w1, w2, w3 = 1, 2, 3
b = 10

y = (x1 * w1) + (x2 * w2) + (x3 * w3) + b
print("La salida de la neurona es ", y)
```
