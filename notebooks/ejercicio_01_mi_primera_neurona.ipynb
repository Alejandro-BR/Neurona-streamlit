# Ejercicio 1 - Una neurona con una entrada

```python
x = input("Introduzca el valor de la entrada: ")
x = float(x)
```

```python
print("y = x * w")
w = 0.5
y = x * w
print(x * w, " = ", x, " * ", w)
print("La salida de la neurona es ", y)
```

# Una neurona con dos entradas

```python
x1, x2 = map(float ,input("Introduzca el valor de los dos entradas separadas por un espacio: ").split())
```

```python
w1, w2 = 1.5, 2.5
```

```python
y = (x1 * w1) + (x2 * w2)
print("La salida de la neurona es ", y)
```
